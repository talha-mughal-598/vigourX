# Ingredients Preprocessing

```python
import nltk
import pandas as pd
import ast
import re
from nltk import word_tokenize, pos_tag
```

```python
recipe = pd.read_csv('/kaggle/input/foodrecsysv1/raw-data_recipe.csv')
```

```python
recipe.head()
```

```python
recipe.ingredients
```

```python
list_of_dict = []

for row in recipe.ingredients:
    list_of_dict.append(row)
```

```python
print(list_of_dict[0])
print(list_of_dict[2])
```

```python
list_ingredients = []

for i in range(len(list_of_dict)):
    list_of_dict[i] = list_of_dict[i].lower()
    list_of_dict[i] = re.split('\^',list_of_dict[i])
    for x in range(len(list_of_dict[i])):
        list_of_dict[i][x] = list_of_dict[i][x].replace('®', '')
        list_of_dict[i][x] = list_of_dict[i][x].replace(':', '')
        list_of_dict[i][x] = re.sub(r'\([^()]*\)', '', list_of_dict[i][x])    
        token = word_tokenize(list_of_dict[i][x])
        tagged = pos_tag(token)
        nouns = [word for word,pos in tagged if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS')]
        list_of_dict[i][x] = ' '.join(nouns)
        if list_of_dict[i][x] not in list_ingredients:
            list_ingredients.append(list_of_dict[i][x])
            list_of_dict[i][x] = len(list_ingredients) - 1
        else:
            list_of_dict[i][x] = list_ingredients.index(list_of_dict[i][x])
#         for j in range(len(list_of_dict[i][x])):
#             if (list_of_dict[i][x][j] < 'A' or (list_of_dict[i][x][j] > 'Z' and list_of_dict[i][x][j] < 'a') or list_of_dict[i][x][j] > 'z') and list_of_dict[i][x][j] != ' ':
#                 print(list_of_dict[i][x][j], end='')
```
